```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
players = pd.read_csv('player_data.csv')
players.head()
```

```python
players['years_play'] = players['year_end'] - players['year_start'] + 1
players.head()
```

```python
players1 = players[['name', 'position', 'years_play']]
players1.sort_values('years_play', ascending=False).head(15)
```

```python
players2 = players[['position', 'years_play']].groupby('years_play').count()
players2.head()
```

```python
#Распределение играков по продолжительсти игры
players2.plot(kind='area')
```

```python
players_season = pd.read_csv('Seasons_Stats.csv')
players_season.head()
```

```python
players_season.info()
```

```python
kobe = players_season[players_season['Player'] == 'Kobe Bryant']
kobe['Year'] = kobe.Year.astype('int64')
kobe
```

```python
# соотношение набранных очков и совершенных бросков
# после получения травмы в 2014 соотношение ухудшилось (приблизительно 1 очко на один бросок)
Kobe_points = kobe[['Year', 'PTS', 'FGA']].plot(kind='bar', x='Year', figsize=(20, 4))
```
