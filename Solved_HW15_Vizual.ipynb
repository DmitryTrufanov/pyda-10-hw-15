```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
players_season = pd.read_csv('Seasons_Stats.csv')
players_season.head()
```

```python
players_season.info()
```

```python
kobe = players_season[players_season['Player'] == 'Kobe Bryant']
kobe['Year'] = kobe.Year.astype('int64')
kobe
```

**столбчатая диаграмма**

```python
# соотношение набранных очков и совершенных бросков
# после получения травмы в 2014 соотношение ухудшилось (приблизительно 1 очко на один бросок)
Kobe_points = kobe[['Year', 'PTS', 'FGA']].plot(kind='bar', x='Year', figsize=(20, 4), title='соотношение набранных очков и выполненных бросков Kobe Bryant')
```

**график**

```python
# процент попадания
Kobe_points2 = kobe[['Year','FG%']].plot(x='Year', figsize=(20, 4), grid=True, title='% точности Kobe Bryant')
# если добавить атрибут kind='area', то изменяется начальная точка оси У на о. Как настроить ось?
```

**круговая диаграмма**

```python
Kobe_skills = kobe[['2P', '3P', 'FT', 'TRB', 'AST', 'STL', 'BLK']].sum()
Kobe_skills.reset_index()
#Kobe_skills.stack().reset_index()
Kobe_skills.plot(kind='pie', y='0', title='Скилы Kobe Bryant')
```

**спарклайн**

```python
#1.группировка
#2.применение функций к столбцам
#3.сортировка
#4.стилизация таблицы - спарклайн

top_15_players = players_season[['Player', 'Year', 'PTS', '2P', '3P', 'FT']].groupby('Player')\
    .agg({'Year': 'count', 'PTS': 'sum', '2P': 'sum', '3P': 'sum', 'FT': 'sum'})\
        .sort_values('PTS', ascending=False).reset_index().head(15)\
            .style.bar('Year') #после применения спарклайна почему-то во всех столбцах увеличилось кол-во знаков после запятой
top_15_players
```

**столбчатая диаграмма с накоплением**

```python
# набор бросков топ 15 играков по количеству точных бросков
top_15_players2 = players_season[['Player', '2P', '3P', 'FT']].groupby('Player')\
    .sum().sort_values('2P', ascending=False).reset_index().head(15)\
            .plot(kind='bar', rot=45, stacked=True, x='Player', figsize=(20, 4), title='Атакующие навыки ТОП-15 играков по набранным очкам')
```

**гистограмма**

```python
players_years = players_season[['Year', 'Player']].groupby('Player').count().sort_values('Year', ascending=False).reset_index()
players_years.head()
```

```python
#Распределение игроков по продолжительсти игры
players_years.hist(figsize=(20, 4), bins=30)

# не соображу как по оси Y расположить лет игры, по X кол-во игроков, настроить плотность подписей оси
```

**точечная диаграмма**

```python
players_position = players_season[['Pos', 'PTS', 'Age']].groupby(['Pos', 'Age']).sum().reset_index()
players_position.head()
```

```python
#распределение количества очков по возрастам играков. Самый результативный возраст 22-27 лет

sns.scatterplot(x='Age', y='PTS',data=players_position)
```
